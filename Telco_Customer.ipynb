```python
import numpy as np
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.columns =df.columns.str.strip().str.lower()
```

```python
#  cleaning the dataset ,converting
# totalcharges to numeric

df['totalcharges']=pd.to_numeric(df['totalcharges'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
# encoding the target variable
df['churn'] = df['churn'].map({'Yes':1,'No':0})
```

```python
# Feature target operation
x = df.drop(['customerid', 'churn'], axis=1)
y = df['churn']

print("Features (X) shape:", x.shape)
print("Target (y) shape:", y.shape)
```

```python
# feature and target separation

x = df.drop(['customerid', 'churn'], axis=1)
y = df['churn']
#  here i removed the unnecessary column customer
# id too
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# scaling in preprocessing is not needed for decision tree model
```

```python
cat_cols = x.select_dtypes(include = 'object').columns
num_cols = x.select_dtypes(exclude = 'object').columns
```

```python
# train test split
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
    )
```

```python
y.value_counts(normalize=True)
```

```python
# building sklearn encoder
preprocessor = ColumnTransformer(
    transformers=[
        ('cat',OneHotEncoder(drop='first',
        handle_unknown='ignore'),cat_cols),
        ('num','passthrough',num_cols)
    ]
)
```

```python
# now we  will create a pipeline end to end for encoder
# one object,no leakage
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
# pipeline = preprocessing + model together
# this avoid the mess up in order

dt_pipeline = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier(random_state=42))
])
```

```python
dt_pipeline.fit(x_train, y_train)
```

```python
# Evaluating the model

from sklearn.metrics import accuracy_score,classification_report
y_pred_dt = dt_pipeline.predict(x_test)
# above one for test
y_pred_train = dt_pipeline.predict(x_train)
# above one for train
print(accuracy_score(y_train,y_pred_train))
print(accuracy_score(y_test,y_pred_dt))

#if train acc is high , test acc is low   overfitting
#if train acc is low lik(50),
# test acc is low - underfitting
# if train and test acc is good enough
# and same - proper
```

```python
print(classification_report(y_test,y_pred_dt))
```

```python
# overfitting intuition
for depth in[2,4,6,8,10,None]:
  dt_pipeline=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier(max_depth=depth,random_state=42))
  ])
  dt_pipeline.fit(x_train,y_train)
  train_acc=dt_pipeline.score(x_train,y_train)
  test_acc=dt_pipeline.score(x_test,y_test)
  print(f'depth:{depth},train={train_acc:.2f},Test={test_acc:.2f}')
```

```python
# feature importance    (model)

dt_model = dt_pipeline.named_steps['model']
encoder = dt_pipeline.named_steps['preprocessor'] # Corrected from 'preprocess'

features_names = encoder.get_feature_names_out()

importance = pd.Series(dt_model.feature_importances_,index=features_names).sort_values(ascending=False)
importance.head(10)

#feature importance
#a pipeline that wraps everything together
#to inspect internals
#dt_model-the trained decision tree
#encode - the fitted ColumnTransformer
#higher value more importances
#values sum to 1.0
#trees calculate this automatically during training
```

```python
# random forest

from sklearn.ensemble import RandomForestClassifier
rf_pipeline = Pipeline([
    ('preprocess',preprocessor),
    ('model',RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        max_depth=6,
        n_jobs=-1
    ))
])
rf_pipeline.fit(x_train,y_train)
```

```python
rf_pipeline.score(x_train,y_train)
```

```python
rf_pipeline.score(x_test,y_test)
```

```python
# gradient boosting
from sklearn.ensemble import GradientBoostingClassifier

gb_pipeline = Pipeline([
    ('preprocess',preprocessor),
    ('model',GradientBoostingClassifier(
    n_estimators = 150,
    learning_rate = 0.1,
    random_state = 42,
    max_depth = 6
    ))
])

gb_pipeline.fit(x_train,y_train)
gb_pipeline.score(x_test,y_test)
```
